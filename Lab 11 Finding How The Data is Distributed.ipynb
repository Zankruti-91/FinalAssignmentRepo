### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head(30)
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
print("Columns in the dataset:")
print(df.columns.tolist())

print("\nData types of each column:")
print(df.dtypes)

print("\nSummary info:")
df.info()

print("\nDescriptive statistics for numeric columns:")
print(df.describe())


print("\nDescriptive statistics for all columns:")
print(df.describe(include='all'))
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
#counting missing value
missing_counts = df.isnull().sum()
print("Missing values per colum:")
print(missing_counts)
```

```python
#percentage of missing values per column
missing_percent = (df.isnull().mean() * 100).round(2)
print("Percentage of missing values per column:")
print(missing_percent)
```

```python
# Numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)
# Categorical columns (usually object or category types)
cat_cols = df.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
#value count employment
print("Employment Distribution:")
print(df['Employment'].value_counts(dropna=False))
print("\n")
#value count jobsat
print("Job Satisfaction Distribution")
print(df["JobSat"].value_counts(dropna=False))
#Value count YearsCodePro
print("Professional Code Experience Distribution:")
print(df['YearsCodePro'].value_counts(dropna=False).sort_index())
#convert to df for plotting or export
employment_dist = df['Employment'].value_counts(dropna=False).reset_index()
employment_dist.columns = ['Employment', 'Count']
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
#drop missing values
job_sat_count = df['JobSat'].value_counts()

#Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(job_sat_count, labels=job_sat_count.index, autopct='%1.1f%%', startangle=140,  colors=plt.cm.Pastel2.colors)
plt.title('Job Satisfaction Distribution')
plt.axis('equal')
plt.show()
```

```python
#You’ll see which satisfaction levels are most common.

#For example:

#If "Very satisfied" and "Satisfied" take up most of the pie, it suggests high overall satisfaction.

#A significant slice for "Dissatisfied" may indicate retention issues or workplace dissatisfaction.
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
#create a copy
df_lang = df[['LanguageHaveWorkedWith','LanguageWantToWorkWith']].dropna().copy()

#split the semicolon-separated strings into lists
df_lang['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].str.split(';')
df_lang['LanguageWantToWorkWith'] = df['LanguageWantToWorkWith'].str.split(';')
```

```python
from collections import Counter

# Flatten the lists and count
have_worked = Counter(lang for sublist in df_lang['LanguageHaveWorkedWith'] for lang in sublist)
want_work = Counter(lang for sublist in df_lang['LanguageWantToWorkWith'] for lang in sublist)
```

```python
all_languages = set(have_worked) | set(want_work)
lang_df = pd.DataFrame({
     'Language': list(all_languages),
    'HaveWorkedWith': [have_worked.get(lang, 0) for lang in all_languages],
    'WantToWorkWith': [want_work.get(lang, 0) for lang in all_languages]
})
```

```python
# Plot top 15 for clarity
top_n = 15
plot_df = lang_df.head(top_n)

plt.figure(figsize=(12, 6))
bar_width = 0.4
x = range(top_n)

plt.bar(x, plot_df['HaveWorkedWith'], width=bar_width, label='Have Worked With', color='skyblue')
plt.bar([i + bar_width for i in x], plot_df['WantToWorkWith'], width=bar_width, label='Want to Work With', color='lightgreen')

plt.xticks([i + bar_width/2 for i in x], plot_df['Language'], rotation=45)
plt.ylabel('Number of Respondents')
plt.title('Programming Languages: Have Worked With vs. Want to Work With')
plt.legend()
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
#group and count values
remote_country = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)
top_countries = remote_country.sum(axis=1).sort_values(ascending=False).head(10).index
plot_data = remote_country.loc[top_countries]

plot_data.plot(kind="bar", figsize=(12, 6), colormap='Set2')

plt.title('Remote Work Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Respondants')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work Type')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
```

```python
corr_df = df[['JobSat', 'YearsCodePro']].dropna()
```

```python
from scipy.stats import pearsonr

pearson_corr, pearson_p = pearsonr(corr_df['JobSat'], corr_df['YearsCodePro'])
print(f"Pearson correlation:{pearson_corr:.3f} (p-value:{pearson_p:.3f})")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
#create the crosstab
employment_edu_ct = pd.crosstab(df['EdLevel'], df['Employment'])
```

```python
#plot
employment_edu_ct.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set3')

plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
