### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# Step 1: Drop null values
df_clean = df.dropna(subset=['DatabaseWantToWorkWith']).copy()

# Step 2: Split semicolon-separated strings into lists
df_clean['DatabaseList'] = df_clean['DatabaseWantToWorkWith'].str.split(';')

# Step 3: Explode the list into rows
df_exploded = df_clean.explode('DatabaseList')

# Remove leading/trailing whitespace
df_exploded['DatabaseList'] = df_exploded['DatabaseList'].str.strip()

# Step 4: Count frequencies of each database
db_counts = df_exploded['DatabaseList'].value_counts()

# Step 5: Select top 5 databases
top5_dbs = db_counts.head(5)

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(
    top5_dbs,
    labels=top5_dbs.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Step 1: Drop nulls in DevType
df_clean = df.dropna(subset=['DevType']).copy()

# Step 2: Split the semicolon-separated developer roles
df_clean['DevTypeList'] = df_clean['DevType'].str.split(';')

# Step 3: Explode the list into rows
df_exploded = df_clean.explode('DevTypeList')

# Remove whitespace
df_exploded['DevTypeList'] = df_exploded['DevTypeList'].str.strip()

# Step 4: Count frequencies
devtype_counts = df_exploded['DevTypeList'].value_counts()

# Step 5: Top 5 developer roles
top5_devtypes = devtype_counts.head(5)

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(
    top5_devtypes,
    labels=top5_devtypes.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Developer Roles')
plt.axis('equal')  # Keeps pie chart circular
plt.show()
```

The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
# Drop missing values in the column
df_clean = df.dropna(subset=['LanguageHaveWorkedWith']).copy()

# Split semicolon-separated values into lists
df_clean['LanguageList'] = df_clean['LanguageHaveWorkedWith'].str.split(';')

# Flatten the list of all languages
all_languages = df_clean['LanguageList'].explode().str.strip()

# Count frequency of each language
language_counts = all_languages.value_counts()

# Take top 5 languages
top5 = language_counts.head(5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Programming Languages Respondents Have Worked With')
plt.axis('equal')  # Makes the pie chart circular
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
# Step 1: Drop null values in the column
df_clean = df.dropna(subset=['NEWCollabToolsHaveWorkedWith']).copy()

# Step 2: Split semicolon-separated values into lists
df_clean['ToolList'] = df_clean['NEWCollabToolsHaveWorkedWith'].str.split(';')

# Step 3: Flatten the list and strip spaces
all_tools = df_clean['ToolList'].explode().str.strip()

# Step 4: Count the frequency of each tool
tool_counts = all_tools.value_counts()

# Step 5: Get top 5 collaboration tools
top5_tools = tool_counts.head(5)

# Step 6: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_tools, labels=top5_tools.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Top 5 Collaboration Tools Used Professionally')
plt.axis('equal')  # Ensures pie is circular
plt.show()
```

### Task 3: Analyzing and Interpreting Composition

##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
# Step 1: Drop rows with null values in LanguageAdmired
df_clean = df.dropna(subset=['LanguageAdmired']).copy()

# Step 2: Split semicolon-separated languages into lists
df_clean['AdmiredLangList'] = df_clean['LanguageAdmired'].str.split(';')

# Step 3: Flatten the lists and strip whitespace
all_languages = df_clean['AdmiredLangList'].explode().str.strip()

# Step 4: Count frequency of each language
lang_counts = all_languages.value_counts()

# Step 5: Get top 5 admired languages
top5_langs = lang_counts.head(5)

# Step 6: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_langs, labels=top5_langs.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Most Admired Programming Languages')
plt.axis('equal')  # Keeps pie circular
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
# Step 1: Drop rows with missing values in AIToolCurrentlyUsing
df_clean = df.dropna(subset=['AIToolCurrently Using']).copy()

# Step 2: Split semicolon-separated tools into individual items
df_clean['AIToolsList'] = df_clean['AIToolCurrently Using'].str.split(';')

# Step 3: Flatten the list and remove extra whitespace
all_tools = df_clean['AIToolsList'].explode().str.strip()

# Step 4: Count each AI tool's usage frequency
tool_counts = all_tools.value_counts()

# Step 5: Get the top 5 most used tools
top5_tools = tool_counts.head(5)

# Step 6: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_tools, labels=top5_tools.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Top 5 AI Tools Currently Used by Developers')
plt.axis('equal')  # Ensures pie is a circle
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
# Step 1: Drop rows with missing values in WebframeWantToWorkWith
df_clean = df.dropna(subset=['WebframeWantToWorkWith']).copy()

# Step 2: Split semicolon-separated frameworks into lists
df_clean['WebFrameworkList'] = df_clean['WebframeWantToWorkWith'].str.split(';')

# Step 3: Flatten list and clean up whitespace
all_frameworks = df_clean['WebFrameworkList'].explode().str.strip()

# Step 4: Count frequency of each framework
framework_counts = all_frameworks.value_counts()

# Step 5: Get top 5 frameworks
top5_frameworks = framework_counts.head(5)

# Step 6: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5_frameworks, labels=top5_frameworks.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Top 5 Preferred Web Frameworks')
plt.axis('equal')  # Ensures pie chart is circular
plt.show()
```
